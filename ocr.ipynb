```python
from cnocr import CnOcr

img_fp = './img/example-chinese-note.jpg'
ocr = CnOcr()  # 所有参数都使用默认值
out = ocr.ocr(img_fp)

print(out)
```

```python
text_sum = ''
for infos in out:
    text_sum+=infos['text']
```
